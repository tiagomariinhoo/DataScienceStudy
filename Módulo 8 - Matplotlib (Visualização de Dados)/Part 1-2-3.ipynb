```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11) #criando um espaço vetorial
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional 
plt.plot(x, y, 'r--') #Com o inline não precisa usar o plt.show()
plt.xlabel('Eixo X') #Texto que vai botar embaixo
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1) #Primeiro argumento: quantas linhas de grafico, proximo argumento col
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Adiciona um eixo
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) #Adiciona um eixo
axes1.plot(x, y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Titulo')
axes2.plot(y, x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--') #Plot do eixo
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titulo')
```

```python
fig = plt.figure(figsize=(8, 4), dpi = 100)
```

```python
fig, axes = plt.subplots(figsize=(8, 5)) #Dimensões da figura

axes.plot(x, y, 'r-', label='x ^ 2')
axes.plot(y, x, 'g--', label='x ** 1/2')
axes.set_title('Titulo')
axes.legend()
```
