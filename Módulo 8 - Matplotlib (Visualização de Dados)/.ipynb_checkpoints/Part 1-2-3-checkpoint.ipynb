```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11) #criando um espa√ßo vetorial
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional 
plt.plot(x, y, 'r--') #Com o inline n√£o precisa usar o plt.show()
plt.xlabel('Eixo X') #Texto que vai botar embaixo
plt.ylabel('Eixo Y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1) #Primeiro argumento: quantas linhas de grafico, proximo argumento col
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Adiciona um eixo
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) #Adiciona um eixo
axes1.plot(x, y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Titulo')
axes2.plot(y, x)
```
